```python
def toh(n, from_danda, to_danda, aux_danda):
    if n == 1:
        print('Move disk 1 from bar', from_danda, "to bar", to_danda)
        return 
    
    toh(n-1, from_danda, aux_danda, to_danda)
    print("Move disk", n, "from bar", from_danda, "to bar", to_danda)
    
    toh(n-1, aux_danda, to_danda, from_danda)
    
```

```python
n = 3

toh(n, 'A', 'C', 'B')
```
